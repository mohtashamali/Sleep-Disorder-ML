```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df)
```

```python
pd.crosstab(df['Age'], df['Sleep Disorder']).plot(kind='bar', figsize=(12, 6))
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Sleep Disorders by Age Group")
plt.legend(title="Sleep Disorder Type")
plt.xticks(rotation=0)
plt.show()
```

```python
pd.crosstab(df['BMI Category'], df['Sleep Disorder']).plot(kind='bar', figsize=(12, 6))
plt.xlabel("BMI Category")
plt.ylabel("Count")
plt.title("Sleep Disorders by BMI Group")
plt.legend(title="Sleep Disorder Type")
plt.xticks(rotation=0)
plt.show()
```
